```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# subplot
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 5);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 5, label = '2024');
plt.grid()
```

```python
# Box plot
```

```python
marks = [67, 65, 55, 59, 69, 72, 51, 70, 53, 98, 23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```
