```python
# square of a number


import tkinter as tk

top = tk.Tk()

top.title('Square')
top.geometry('500x400')

a = tk.IntVar()

def square():
    num = a.get()
    sq = num * num
    result = tk.Label(top, text='Square is %i' % sq).place(x=120, y=200)

l = tk.Label(top, text='INPUT').place(x=20, y=100)
e = tk.Entry(top, textvariable=a).place(x=100, y=100)
b = tk.Button(top, text='Submit', bg='orange', command=square).place(x=120, y=170)

top.mainloop()
```

```python
# square of a number

import tkinter as tk
from tkinter import messagebox

top = tk.Tk()

top.title('Square')
top.geometry('500x400')

a = tk.IntVar()

def square():
    num = a.get()
    sq = num * num
    result = tk.Label(top, text='Square is %i' % sq).place(x=120, y=200)
    messagebox.showinfo("Result", "Square is %i" % sq)

l = tk.Label(top, text='INPUT').place(x=20, y=100)
e = tk.Entry(top, textvariable=a).place(x=100, y=100)
b = tk.Button(top, text='Submit', bg='orange', command=square).place(x=120, y=150)

top.mainloop()
```

```python
# create for Addition, sub, mult, div of 2 numbers
# Addition

import tkinter as tk
from tkinter import messagebox

top = tk.Tk()

top.title('Addition')
top.geometry('500x400')

a1 = tk.IntVar()
a2 = tk.IntVar()

def add():
    ad = a1.get() + a2.get()
    messagebox.showinfo("Result", "Addition is %i" % ad)

l = tk.Label(top, text='Input 1:').place(x=20, y=100)
l1 = tk.Label(top, text='Input 2:').place(x=20, y=130)
e = tk.Entry(top, textvariable=a1).place(x=100, y=100)
e1 = tk.Entry(top, textvariable=a2).place(x=100, y=130)

b = tk.Button(top, text='Submit', bg='orange', command=add, cursor='hand2').place(x=120, y=150)

top.mainloop()
```

```python
import tkinter as tk
from tkinter import StringVar, Button, Entry

top = tk.Tk()
top.title('Calculator')

# Define the StringVar for the display
display = StringVar()
display.set("0")

# Create an Entry widget for the display
entry = Entry(top, textvariable=display, width=16, font=('Arial', 24), bd=8, insertwidth=2, justify='right')
entry.grid(row=0, column=0, columnspan=4)

# Function to handle button clicks
def button_click(char):
    current = display.get()
    if current == "0":
        display.set(char)
    else:
        display.set(current + char)

# Function to clear the display
def button_clear():
    display.set("0")

# Function to evaluate the expression
def button_equal():
    try:
        result = str(eval(display.get()))
        display.set(result)
    except:
        display.set('Error')

# Create buttons
button_7 = Button(top, text="7", command=lambda: button_click("7"), width=5, height=2)
button_7.grid(row=1, column=0)
button_8 = Button(top, text="8", command=lambda: button_click("8"), width=5, height=2)
button_8.grid(row=1, column=1)
button_9 = Button(top, text="9", command=lambda: button_click("9"), width=5, height=2)
button_9.grid(row=1, column=2)

button_div = Button(top, text="Division", command=lambda: button_click("/"), width=5, height=2)
button_div.grid(row=1, column=3)
button_4 = Button(top, text="4", command=lambda: button_click("4"), width=5, height=2)
button_4.grid(row=2, column=0)
button_5 = Button(top, text="5", command=lambda: button_click("5"), width=5, height=2)
button_5.grid(row=2, column=1)
button_6 = Button(top, text="6", command=lambda: button_click("6"), width=5, height=2)
button_6.grid(row=2, column=2)

button_multi = Button(top, text="Multi", command=lambda: button_click("*"), width=5, height=2)
button_multi.grid(row=2, column=3)
button_1 = Button(top, text="1", command=lambda: button_click("1"), width=5, height=2)
button_1.grid(row=3, column=0)
button_2 = Button(top, text="2", command=lambda: button_click("2"), width=5, height=2)
button_2.grid(row=3, column=1)
button_3 = Button(top, text="3", command=lambda: button_click("3"), width=5, height=2)
button_3.grid(row=3, column=2)

button_sub = Button(top, text="Substract", command=lambda: button_click("-"), width=5, height=2)
button_sub.grid(row=3, column=3)
button_0 = Button(top, text="0", command=lambda: button_click("0"), width=5, height=2)
button_0.grid(row=4, column=0)
button_dot = Button(top, text=".", command=lambda: button_click("."), width=5, height=2)
button_dot.grid(row=4, column=1)
button_clear = Button(top, text="Clear", command=button_clear, width=5, height=2)
button_clear.grid(row=4, column=2)

button_add = Button(top, text="ADD", command=lambda: button_click("+"), width=5, height=2)
button_add.grid(row=4, column=3)
button_equal = Button(top, text="=", command=button_equal, width=22, height=2)
button_equal.grid(row=5, column=0, columnspan=4)

top.mainloop()
```
