```python
sns.boxplot(df['price']);
plt.title('outliers in price');
```

```python
def ol(data):
    q3=np.percentile(data,75)
    q1=np.percentile(data,25)
    
    iqr=q3-q1
    
    ub=q3+(1.5*iqr)
    lb=q1-(1.5*iqr)
    
    return ub,lb
```

```python
ub,lb=ol(df['price'])
```

```python
# filtering the data without outliers
df=df[(df['price']<=ub)&(df['rating']>=lb)]
```

```python
sns.boxplot(df['price']);
plt.title('outliers in price');
```
