```python
def ol(data):
    q3=np.percentile(data,75)
    q1=np.percentile(data,25)
    
    iqr=q3-q1
    
    ub=q3+(1.5*iqr)
    lb=q1-(1.5*iqr)
    
    return ub,lb
```

```python
ub,lb=ol(df['price'])
```

```python
# filtering the data without outliers
df=df[(df['price']<=ub)&(df['rating']>=lb)]
```

```python
# Drop rows with missing values in the columns of interest
data_clean = data.dropna(subset=['price', 'rating'])

# Scale the price values for better readability (optional)
data_clean['price_scaled'] = data_clean['price'] / 1000  # Represent prices in thousands

# Aggregate the data to calculate the average price for each rating
average_price_by_rating = data_clean.groupby('rating')['price_scaled'].mean().reset_index()

# Define x and y for the plot
x = average_price_by_rating['rating']
y = average_price_by_rating['price_scaled']

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', mfc='red', color='black')

plt.title('Average Price (in Thousands) by Rating')
plt.xlabel('Rating')
plt.ylabel('Average Price (in Thousands)')
plt.grid(True)
plt.show()
```

```python
plt.title('Average Price (in Thousands) by Rating')
plt.xlabel('Rating')
plt.ylabel('Average Price (in Thousands)')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker='o', mfc = 'red', color = 'black', ls = '--', ms = 5, label = '2023')
plt.plot(x, y, marker='o', mfc = 'red', color = 'black', ls = '--', ms = 5, label = '2024')
plt.grid()
plt.legend()
```

```python
brand_counts = data['brand_name'].value_counts()
threshold = 20  # Minimum count to be shown separately
other_count = brand_counts[brand_counts < threshold].sum()

brand_counts = data['brand_name'].value_counts()
brand_counts_filtered = brand_counts[brand_counts >= threshold]
brand_counts_filtered['Other'] = other_count
```

```python
# Plotting the pie chart

plt.figure(figsize=(10, 6))
plt.pie(brand_counts_filtered, labels=brand_counts_filtered.index, autopct='%1.1f%%', startangle=140, )
plt.title('Distribution of Smartphones by Brand')
```

```python
# plotting the Bar plot

plt.figure(figsize=(12, 6))
brand_counts.plot(kind='bar', color='Green')
plt.title('Distribution of Smartphones by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Smartphones')
```
